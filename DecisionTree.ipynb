```python
import utils.reader as reader
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
from sklearn import metrics 
from sklearn import tree
from sklearn.externals import joblib
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x_train, y_train = reader.load_mnist('data/fashion', kind='train')
x_test, y_test = reader.load_mnist('data/fashion', kind='t10k')

print(x_train.shape)
print(x_test.shape)

x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
recogniser = joblib.load('filename2.pkl')
```

```python
print('\nMaking Predictions on Validation Data...')
y_pred = recogniser.predict(x_test)
print(y_pred)
```

```python
print('\nCalculating Accuracy of trained Classifier...')
confidence = metrics.accuracy_score(y_test, y_pred)
print(confidence)
```

```python
print('\nCreating Confusion Matrix...')
conf_mat = confusion_matrix(y_test,y_pred)

df_cm = pd.DataFrame(conf_mat, index = [i for i in "0123456789"],
                  columns = [i for i in "0123456789"])
plt.figure(figsize = (10,7))
plt.ylabel('True label')
plt.xlabel('Predicted label')
sn.heatmap(df_cm, annot=True)
```
